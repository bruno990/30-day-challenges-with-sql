````python
%%sql

SELECT * FROM Customer LIMIT 10
````

````python
%%sql result_set << 

SELECT (CustomerId) as codigo, (FirstName || ' ' || lower(LastName)) nome, (lower(Address)) as endereco, (lower(City) || '-' || upper(State)) as cidade, (upper(Country)) as pais,
    (PostalCode) as cep, replace(Phone, '^+', '') as telefone, Email 
FROM Customer 
WHERE pais = 'BRAZIL'
````
